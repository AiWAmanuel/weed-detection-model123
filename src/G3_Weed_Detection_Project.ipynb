```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import img_to_array, load_img
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models

# ------------------------ 1. Data Preprocessing ------------------------

# Set up directories for dataset (replace with your actual directory)
weed_dir = 'C:\\Users\\Hp\\Desktop\\VSCode\\G3_Weed_Detection_Project\\archive\\train'  # Weed images directory
non_weed_dir = 'C:\\Users\\Hp\\Desktop\\VSCode\\G3_Weed_Detection_Project\\archive\\test'  # Non-weed images directory

# Initialize variables for images and labels
images = []
labels = []

# Function to load and process images from a given directory
def load_images_from_directory(directory, label):
    """
    Loads images from the specified directory, resizes them to 128x128 pixels,
    and appends them to the images list with corresponding labels.
    """
    for img_name in os.listdir(directory):
        img_path = os.path.join(directory, img_name)
        
        # Skip non-image files (e.g., corrupted or unsupported files)
        try:
            # Load image with PIL (Pillow), resize it to 128x128
            img = load_img(img_path, target_size=(128, 128))
            
            # Convert image to a numpy array and normalize pixel values to range 0-1
            img_array = img_to_array(img) / 255.0  # Normalize pixel values to [0, 1]
            
            # Append the image array and corresponding label
            images.append(img_array)
            labels.append(label)  # 1 for weed, 0 for non-weed
        except (OSError, IOError, ValueError):
            # Skip files that are not valid image files or cannot be opened
            print(f"Skipping non-image file: {img_name}")

# Load images from both directories
load_images_from_directory(weed_dir, 1)  # Label weed images as 1
load_images_from_directory(non_weed_dir, 0)  # Label non-weed images as 0

# Convert lists to numpy arrays for easier manipulation with TensorFlow
images = np.array(images)
labels = np.array(labels)

# Split the dataset into training and validation sets (80% training, 20% validation)
X_train, X_val, y_train, y_val = train_test_split(images, labels, test_size=0.2, random_state=42)

# ------------------------ 2. Model Building ------------------------

# Build a Convolutional Neural Network (CNN)
model = models.Sequential()

# Input Layer: The first layer expects input images of size (128, 128, 3)
model.add(layers.InputLayer(input_shape=(128, 128, 3)))

# Convolutional Layer 1: Apply 32 filters, kernel size (3x3), ReLU activation
model.add(layers.Conv2D(32, (3, 3), activation='relu'))

# Pooling Layer 1: Max-pooling to reduce dimensionality
model.add(layers.MaxPooling2D(pool_size=(2, 2)))

# Convolutional Layer 2: Apply 64 filters, kernel size (3x3), ReLU activation
model.add(layers.Conv2D(64, (3, 3), activation='relu'))

# Pooling Layer 2: Max-pooling to reduce dimensionality further
model.add(layers.MaxPooling2D(pool_size=(2, 2)))

# Convolutional Layer 3: Apply 128 filters, kernel size (3x3), ReLU activation
model.add(layers.Conv2D(128, (3, 3), activation='relu'))

# Pooling Layer 3: Max-pooling to reduce dimensionality further
model.add(layers.MaxPooling2D(pool_size=(2, 2)))

# Flatten the 3D outputs of the convolutional layers to 1D
model.add(layers.Flatten())

# Fully Connected Layer: Dense layer to process the extracted features
model.add(layers.Dense(128, activation='relu'))

# Output Layer: A single neuron for binary classification with sigmoid activation
model.add(layers.Dense(1, activation='sigmoid'))

# Print the model summary to see the architecture
model.summary()

# ------------------------ 3. Model Training ------------------------

# Compile the model
model.compile(optimizer='adam',  # Adam optimizer for faster convergence
              loss='binary_crossentropy',  # Binary cross-entropy loss for binary classification
              metrics=['accuracy'])  # Track accuracy during training

# Train the model on the training data
history = model.fit(X_train, y_train, 
                    batch_size=32,  # Batch size for training
                    epochs=10,  # Number of training epochs
                    validation_data=(X_val, y_val))  # Use validation data for evaluation

# ------------------------ 4. Model Evaluation ------------------------

# Evaluate the model on the validation data
val_loss, val_accuracy = model.evaluate(X_val, y_val)
print(f"Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_accuracy:.4f}")

# Optional: Plot training and validation accuracy and loss
plt.figure(figsize=(12, 6))

# Plot Accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Plot Loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()

# ------------------------ 5. Prediction and Output ------------------------

# Predict class for new images (example usage)
def predict_image(img_path):
    """
    Given a file path of an image, preprocesses it and makes a prediction using the trained model.
    """
    img = load_img(img_path, target_size=(128, 128))  # Resize image to match input size
    img_array = img_to_array(img) / 255.0  # Normalize pixel values to [0, 1]
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension

    # Predict class (weed or non-weed)
    prediction = model.predict(img_array)
    return 'Weed' if prediction >= 0.5 else 'Non-Weed'

# Example prediction on a new image (replace with your image path)
prediction = predict_image('C:/Users/Hp/Desktop/VSCode/G3_Weed_Detection_Project/archive/test/1.jpg')

print(f"Prediction for the image: {prediction}")

# ------------------------ 6. Model Saving ------------------------

# Save the trained model for future use
model.save('weed_detection_model.h5')  # Save the model as an .h5 file

# Optionally, you can use OpenCV or other libraries to visualize predictions on images
```
